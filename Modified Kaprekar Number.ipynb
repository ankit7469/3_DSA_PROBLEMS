```python
# Modified Kaprekar Number

p = int(input())
q = int(input())

found= False
for i in range(p, q+1):

    m = i*i
    s = str(m)

    left = int(s[:len(s)//2] or 0)
    right = int(s[len(s)//2:])

    n=left + right 
    
    if n == i:
        print(i,end=" ")
        found=True
        
if not found:
    print("INVALID RANGE")
```
