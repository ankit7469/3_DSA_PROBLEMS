```python
# Lisa's Workbook 

n,k = map(int,input().split())  
arr = list(map(int,input().split()))

count = 0
page = 1

for i in arr:                                          #  chapter
    problem = 1                                        # problem start
    
    while problem <= i:                                # until problem ends
        
        start = problem
        end = min(problem + k - 1, i)
        
        # check special problem
        if page >= start and page <= end:
            count += 1
        
        problem += k
        page += 1

print(count)
```

```python
Sample Input

STDIN       Function
-----       --------
5 3         n = 5, k = 3
4 2 6 1 10  arr = [4, 2, 6, 1, 10]
Sample Output

4
```

```python
# Short version
n,k = map(int,input().split())
arr = list(map(int,input().split()))
page = count = 0

for x in arr:
    for i in range(1, x+1, k):
        page += 1
        if page in range(i, min(i+k, x+1)):
            count += 1

print(count)
```
