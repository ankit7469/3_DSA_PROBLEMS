```python
# Flatland Space Station

n, m = map(int,input().split())
c = list(map(int,input().split()))

c.sort()

max_dist = 0

max_dist = max(max_dist, c[0] - 0)       # Distance between first city and first station

for i in range(len(c)-1):                # Max mid point between stations   , len(c)==2 (0,4)
    dist = (c[i+1] - c[i]) // 2          # i+1=4 and i=0
    max_dist = max(max_dist, dist)

max_dist = max(max_dist, (n-1) - c[-1])  # distance between last station to last city

print(max_dist)
```
