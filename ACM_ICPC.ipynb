```python
# ACM ICPC

n, m = map(int, input().split())

topics = []

for _ in range(n):
    topics.append(input())

max_topics = 0
team_count = 0

for i in range(n):
    for j in range(i + 1, n):

        count = 0

        for k in range(m):
            if topics[i][k] == '1' or topics[j][k] == '1':
                count += 1

        if count > max_topics:
            max_topics = count
            team_count = 1
        elif count == max_topics:
            team_count += 1

print(max_topics)
print(team_count)
```
