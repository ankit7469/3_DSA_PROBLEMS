```python
## Organizing containers of balls

t = int(input())

for _ in range(t):
    n = int(input())
    matrix = []
    
    for i in range(n):
        row = list(map(int, input().split()))
        matrix.append(row)
    
    
    row_sum = []                        # Row sums (container capacity)
    for i in range(n):
        row_sum.append(sum(matrix[i]))
    
    col_sum = []                       # Column sums (ball type total)
    for j in range(n):
        total = 0
        for i in range(n):
            total += matrix[i][j]
        col_sum.append(total)
    
# Sort and compare
    
    if sorted(row_sum) == sorted(col_sum):
        print("Possible")
    else:
        print("Impossible")
```
